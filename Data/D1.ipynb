```python
#Import závislostí
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Načtení dat
steam_df = pd.read_csv("steam.csv")
```

```python
# Zobrazení základních informací
display(steam_df.head())
steam_df.info()
steam_df.describe()
```

```python
##Počet prázdných hodnot
steam_df.isnull().sum().sort_values(ascending=False)
```

```python
## Distribuce cen
plt.figure(figsize=(10, 5))
steam_df['price'].dropna().astype(float).plot.hist(bins=50)
plt.title("Distribuce cen her na Steamu")
plt.xlabel("Cena ($)")
plt.ylabel("Počet her")
plt.grid(True)
plt.show()
```

```python
# Korelace cena x hodnocení
steam_df['total_ratings'] = steam_df['positive_ratings'] + steam_df['negative_ratings']
steam_df['rating_ratio'] = steam_df['positive_ratings'] / steam_df['total_ratings']

correlation_matrix = steam_df[['price', 'rating_ratio']].corr()

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(5, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("Korelační matice: Cena vs. Poměr hodnocení")
plt.show()
```

```python
## Nejčastější vydavatelé
steam_df['publisher'].value_counts().head(10).plot(kind='barh', figsize=(10, 5), title="Top 10 vydavatelů")
plt.xlabel("Počet her")
plt.show()
```

```python
# Korelační matice
steam_df['total_ratings'] = steam_df['positive_ratings'] + steam_df['negative_ratings']
steam_df['rating_ratio'] = steam_df['positive_ratings'] / steam_df['total_ratings']
steam_df['negative_ratio'] = steam_df['negative_ratings'] / steam_df['total_ratings']

corr_df = steam_df[['price', 'rating_ratio', 'negative_ratio', 'average_playtime', 'median_playtime', 'required_age']]
correlation_matrix = corr_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Korelační matice mezi vybranými proměnnými (relativní hodnocení)")
plt.show()
```

```python
## Podíl platforem
steam_df['windows'] = steam_df['platforms'].str.contains('windows')
steam_df['mac'] = steam_df['platforms'].str.contains('mac')
steam_df['linux'] = steam_df['platforms'].str.contains('linux')

platform_counts = steam_df[['windows', 'mac', 'linux']].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 8))
platform_counts.plot.pie(autopct='%1.1f%%', startangle=90)
plt.ylabel('')
plt.title("Podíl platforem mezi hrami")
plt.show()
```

```python
## Distribuce počtů achievementů
# Filtrování extrémních hodnot
filtered_achievements = steam_df[steam_df['achievements'] < 200]

plt.figure(figsize=(10, 5))
filtered_achievements['achievements'].plot.hist(bins=100)
plt.title("Distribuce počtu achievementů")
plt.xlabel("Počet achievementů")
plt.ylabel("Počet her")
plt.xticks(range(0, 201, 25))
plt.grid(True)
plt.show()
```
